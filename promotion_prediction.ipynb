```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python


# Assuming emp_data and job_performance_data are your dataframes
# Merge both dataframes on 'emp_id'
emp_job_performance = pd.merge(emp_data, job_performance_data, on='emp_id')

```

```python
# Filter promoted employees
promoted_data = emp_job_performance[emp_job_performance['promotion'] == 1]

# Group by designation to count the number of promotions per designation
designation_promotion_counts = promoted_data.groupby('designation')['promotion'].count().reset_index()

# Set up the plot size and style
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Create a bar plot using seaborn
bar_plot = sns.barplot(x='designation', y='promotion', data=designation_promotion_counts, palette='Blues_d')

# Add title and labels
plt.title('Promotions by Designation', fontsize=16)
plt.xlabel('Designation', fontsize=14)
plt.ylabel('Number of Promotions', fontsize=14)

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Add data labels on top of the bars
for index, row in designation_promotion_counts.iterrows():
    bar_plot.text(index, row['promotion'] + 0.1, round(row['promotion'], 2), color='black', ha="center", fontsize=12)

# Show the plot with tight layout
plt.tight_layout()
plt.show()
```

```python
promoted_data = emp_job_performance[emp_job_performance['promotion'] == 1]

# Calculate the total number of promoted employees
total_promoted = len(promoted_data)

# Calculate the number of promoted employees above and below 30
above_30_promoted = len(promoted_data[promoted_data['age'] > 30])
below_30_promoted = len(promoted_data[promoted_data['age'] <= 30])

# Calculate percentages
above_30_percentage = (above_30_promoted / total_promoted) * 100
below_30_percentage = (below_30_promoted / total_promoted) * 100

# Prepare data for plotting
age_groups = ['Above 30', '30 or Below']
percentages = [above_30_percentage, below_30_percentage]

# Set up the plot size and style
plt.figure(figsize=(8, 5))
sns.set_style("whitegrid")

# Create a bar plot using seaborn
sns.barplot(x=age_groups, y=percentages, palette='Blues_d')

# Add title and labels
plt.title('Percentage of Promoted Employees by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Percentage of Promoted Employees (%)', fontsize=14)

# Add data labels on top of the bars
for index, value in enumerate(percentages):
    plt.text(index, value + 1, f'{value:.2f}%', color='black', ha='center', fontsize=12)

# Show the plot with tight layout
plt.tight_layout()
plt.show()
```

```python
emp_skill_job_performance = pd.merge(emp_skill_data, job_performance_data, on='emp_id')
```

```python
# Filter promoted employees
promoted_data = emp_skill_job_performance[emp_skill_job_performance['promotion'] == 1]

# Group by skill_name to count the number of promotions per skill
skill_promotion_counts = promoted_data.groupby('skill_name')['promotion'].count().reset_index()

# Set unique colors for the pie chart
num_skills = len(skill_promotion_counts)
colors = plt.cm.tab20(np.linspace(0, 1, num_skills))  # Use a colormap with unique colors

# Set up the pie chart
plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(skill_promotion_counts['promotion'], 
                                    labels=skill_promotion_counts['skill_name'], 
                                    autopct='%1.1f%%', 
                                    startangle=140,
                                    colors=colors)

# Add a title
plt.title('Percentage of Promotions by Skill', fontsize=16)

# Show the legend
plt.legend(wedges, skill_promotion_counts['skill_name'], title='Skills', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

# Show the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# Calculate total employees and promoted employees for each skill
skill_counts = emp_skill_job_performance.groupby('skill_name').agg(
    total_employees=('emp_id', 'count'),
    promoted_employees=('promotion', 'sum')
).reset_index()

# Calculate promotion rate for each skill
skill_counts['promotion_rate'] = (skill_counts['promoted_employees'] / skill_counts['total_employees']) * 100

# Set up the plot size and style
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Create a bar plot using seaborn
sns.barplot(x='promotion_rate', y='skill_name', data=skill_counts.sort_values('promotion_rate', ascending=False), palette='Blues_d')

# Add title and labels
plt.title('Promotion Rate by Skill', fontsize=16)
plt.xlabel('Promotion Rate (%)', fontsize=14)
plt.ylabel('Skill', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```
